# A notebook for extracting random images. No need for projection. Suitable for a CycleGAN training.

```python
# Load in the libraries
import rasterio as rio
import numpy as np
from rasterio.mask import mask
from rasterio.plot import show
from matplotlib import pyplot as plt
```

```python
# Path to the file
fp = r'/mnt/veracrypt2/CGRE/raw/Mars/ESP_073873_1680_MRGB.JP2'

# Load it into rasterio:

sat = rio.open(fp)

# Display the image

show(sat)
    
```

```python
# Plot the RGB channels

fig, (axr, axg, axb) = plt.subplots(1,3, figsize = (21, 7))

show((sat, 1), ax = axr, cmap = 'Reds', title = 'Red channel')

show((sat, 2), ax = axg, cmap = 'Greens', title = 'Green channel')

show((sat, 3), ax = axb, cmap = 'Blues', title = 'Blue channel')
```
